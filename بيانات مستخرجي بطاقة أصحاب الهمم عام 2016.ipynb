# بيانات مستخرجي بطاقة أصحاب الهمم عام 2016

```python
import pandas as pd
```

```python
df = pd.read_excel("pod_card_db.xlsx")
pd.set_option("display.max_rows", 999)
```

```python
df.head()
```

```python
df[df.duplicated()]
df.isna().sum()
```

# الجنس

```python
gender = df['Gender'].value_counts().rename('العدد').to_frame()
gender.loc['المجموع'] = gender['العدد'].sum()
gender['النسبة'] = (gender['العدد']/len(df)*100).round(1)
gender.columns.name= 'الجنس'
gender
```

# عام الميلاد

```python
birth = df['Year of birth'].value_counts().rename('العدد').to_frame()
birth.sort_index(ascending=True,inplace=True)
birth.loc['المجموع'] = birth['العدد'].sum()
birth.columns.name = 'عام الميلاد'
birth
```

# نوع الإعاقة

```python
disability = df['type of disability'].value_counts().rename('العدد').to_frame()
disability.loc['المجموع'] = disability['العدد'].sum()
disability['النسبة'] = (disability['العدد']/len(df)*100).round(2)
disability.columns.name = 'نوع الإعاقة'
disability
```

# الإمارة

```python
emirate = df['Emirate'].value_counts().rename('العدد').to_frame()
emirate.loc['المجموع'] = emirate['العدد'].sum() 
emirate['النسبة'] = (emirate['العدد']/len(df)*100).round(2)
emirate.columns.name = 'الإمارة'
emirate
```

# مواطن/مقيم

```python
citizens_noncitizens = df['Citizens-Non-Citizens'].value_counts().rename('العدد').to_frame()
citizens_noncitizens.loc['المجموع'] = citizens_noncitizens['العدد'].sum() 
citizens_noncitizens['النسبة'] = ((citizens_noncitizens['العدد']/len(df))*100).round(2)
citizens_noncitizens
```

# الجنسية

```python
nationality = df['nationality'].value_counts().rename('العدد').to_frame()
nationality.loc['المجموع'] = nationality['العدد'].sum() 
nationality['النسبة']= ((nationality['العدد']/len(df))*100).round(2)
nationality.columns.name = 'الجنسية'
nationality
```

أكثر أعوام ميلاد مستخرجي البطاقات حسب الجنس

:الذكور

```python
gender_birth = df.pivot_table(index ='Year of birth' , columns='Gender' , aggfunc ='size' ).fillna(0).astype(int)
gender_birth.index.name = 'عام الميلاد'
gender_birth.sort_values(by='ذكر',ascending=False,inplace=True) 
gender_birth['ذكر'].head(10).to_frame()
```

:الإناث

```python
gender_birth.sort_values(by='أنثى',ascending=False,inplace=True) 
gender_birth['أنثى'].head(10).to_frame()
```

أكثر أعوام ميلاد مستخرجي البطاقات حسب المواطنين والمقيمين

:المواطنين

```python
birth = df.pivot_table(index ='Year of birth', columns='Citizens-Non-Citizens', aggfunc ='size').fillna(0).astype(int)
birth.index.name = 'عام الميلاد'
birth.sort_values(by='مواطن',ascending=False,inplace=True)
birth['مواطن'].head(10).to_frame()
```

:المقيمين

```python
birth.sort_values(by='غير مواطن',ascending=False,inplace=True)
birth['غير مواطن'].head(10).to_frame()
```

أكثر أعوام ميلاد مستخرجي البطاقات حسب الإمارات

```python
birth_emirate = df.pivot_table(index ='Year of birth' , columns='Emirate' , aggfunc ='size' ).fillna(0).astype(int)

lst = ['أبوظبي','الشارقة']
for i in lst:
    top = birth_emirate.sort_values(by=i,ascending=False).head(2)
    print(f"{i}\n")
    for i in top.index:
        print(i)
    print(f"{'ـ'*15}\n")
    
lst = ['أم القيوين', 'الفجيرة', 'دبي', 'رأس الخيمة','عجمان']
for i in lst:
    top = birth_emirate.sort_values(by=i,ascending=False).head(3)
    print(f"{i}\n")
    for i in top.index:
        print(i)
    print(f"{'ـ'*15}\n")
```
